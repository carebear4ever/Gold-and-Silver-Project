```python
import os 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import hvplot.pandas
```

```python
# set file path for Gold
csvpath = Path("./Rice Work Folder/resources project01/gold_price_yearly.csv")
```

```python
#reading csv file for gold's yearly price 
gold_prices = pd.read_csv(csvpath)
```

```python
# display first and last 5 rows of data
display(gold_prices.head())
```

```python
# 'set_index' to set the `Year` as the index to the DataFrame
gold_prices.set_index('Year', inplace=True)
```

```python
# specify range of dates for 10 years using '.loc'
gold_prices = gold_prices.loc['2012':'2022', :]
gold_prices
```

```python
# drop extra columns
gold_prices.drop(columns=['Year Open', 'Year High', 'Year Low', 'Year Range Price', 'Annual % Change'], inplace=True)
gold_prices.head()
```

```python
# plot 2012-2022
gold_prices.hvplot.line(title='Gold Price Over 10 Years')
```

```python
#read silver csv file path
silver_prices = pd.read_csv(
    Path("./Rice Work Folder/resources project01/silver.csv"), 
    index_col="Year"
)
```

```python
# display fist and last 5 rows of silver's dataframe
display(silver_prices.head())
```

```python
# drop extra columns
silver_prices.drop(columns=['Year Open', 'Year High', 'Year Low', 'Annual\n% Change'], inplace=True)
silver_prices.head()
```

```python
#plot 2012 - 2022
silver_prices['Average\nClosing Price'] = silver_prices['Average\nClosing Price'].str.replace('$', '').astype(float)
silver_prices['Year Close'] = silver_prices['Year Close'].str.replace('$', '').astype(float)
silver_prices = silver_prices.sort_values(by='Year')

silver_prices.hvplot.line(x='Year', y=['Average\nClosing Price', 'Year Close'], title='Silver Prices Over 10 Years')
```

```python
# set file path for S&P 500
csvpath = Path("./Rice Work Folder/resources project01/SPX.csv")
```

```python
# reading csv file for S&P 500
SP_500 = pd.read_csv(csvpath)
```

```python
# display first and last 5 rows of dataframe
display(SP_500.head())
```

```python
SP_500.dtypes
```

```python
#transform date column to a datetime data type
SP_500["Date"] = pd.to_datetime(
    SP_500["Date"],
    infer_datetime_format =True,
    utc = True
)
```

```python
# 'set_index' to set the `Year` as the index to the DataFrame
SP_500.set_index('Date', inplace=True)
```

```python
# specify range of dates for 10 years using '.loc'
SP_500 = SP_500.loc['2010-01-04':'2020-11-04', :]
display(SP_500.head())
```

```python
# drop extra columns
SP_500.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'], inplace=True)
display(SP_500.head())
display(SP_500.tail())
```

```python
# plot 2010-2020
SP_500.hvplot.line(
    x='Date',
    y='Close',
    rot=90,
    title='S&P 500 Daily Returns'
)
```

```python
#rolling 180-day mean of S&P500
SP_500.rolling(window=180).mean().plot()
```

```python
#rolling 180-day std dev of S&P 500
SP_500.rolling(window=180).std().plot()
```

```python
#Plot overlay daily closing price and 180-Day rolling mean of S&P 500
#set figure of daily closing prices of S&P 500
ax = SP_500.plot()
#180-Day rolling mean
SP_500.rolling(window=180).mean().plot(ax=ax)
#set legend
ax.legend(["S&P 500 STD", "S&P 500 180 Day Mean"]);
```

```python
#concate gold and silver 
gold_silver_combined = pd.concat([gold_prices, silver_prices], axis="columns", join="inner")
gold_silver_combined
```

```python
gold_silver_combined.drop(columns=['Average Closing Price', 'Average\nClosing Price'], inplace=True)
gold_silver_combined
```

```python
gold_silver_combined.columns = ['Gold', 'Silver']
display(gold_silver_combined.head())
```

```python
csvpath = Path("./Rice Work Folder/resources project01/gold_price_yearly.csv")
gold_prices = pd.read_csv(csvpath)
gold_prices.set_index('Year', inplace=True)
gold_prices = gold_prices.loc['2012':'2022', :]
gold_prices.drop(columns=['Year Open', 'Year High', 'Year Low', 'Year Range Price', 'Annual % Change', 'Average Closing Price'], inplace=True)
silver_prices = pd.read_csv(
    Path("./Rice Work Folder/resources project01/silver.csv"), 
    index_col="Year"
)
silver_prices['Year Close'] = silver_prices['Year Close'].str.replace('$', '').astype(float)
silver_prices.drop(columns=['Year Open', 'Year High', 'Year Low', 'Annual\n% Change', 'Average\nClosing Price'], inplace=True)
gold_prices.hvplot.bar(label="Gold") * silver_prices.hvplot.bar(label="Silver")
```
